**Dataset Description:**

The dataset used in this assignment consists of two CSV files containing information about IPL matches and ball-by-ball deliveries from 2008 to 2019.

• matches.csv – Contains match-level information such as teams, venue, toss result, match outcome, and player of the match.

```python
# Max and Min margins [cite: 65, 66, 67, 68]
print("Max runs win:", m.iloc[m['win_by_runs'].idxmax()]['winner'])
print("Max wickets win:", m.iloc[m['win_by_wickets'].idxmax()]['winner'])
print("Min runs win:", m[m['win_by_runs']>0]['win_by_runs'].min())
print("Min wickets win:", m[m['win_by_wickets']>0]['win_by_wickets'].min())
```

```python
# Counts per category [cite: 70, 71, 72]
m['city'].value_counts()
m['winner'].value_counts()
m['season'].value_counts().plot(kind='bar', title='Matches per Season')
plt.show()
```

```python
# Top 10 Players [cite: 73]
m['player_of_match'].value_counts().head(10).plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 10 Match Winners')
plt.show()
```

```python
# Toss decision vs Result [cite: 74, 75]
print("Toss decision counts:\n", m['toss_decision'].value_counts())
toss_win_match_win = (m['toss_winner'] == m['winner']).value_counts()
sns.barplot(x=toss_win_match_win.index, y=toss_win_match_win.values)
plt.title('Does Toss Win = Match Win?')
plt.show()
```
